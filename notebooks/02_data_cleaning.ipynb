```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv("../data/olist_customers_dataset.csv")
orders = pd.read_csv("../data/olist_orders_dataset.csv")
products = pd.read_csv("../data/olist_products_dataset.csv")
payments = pd.read_csv("../data/olist_order_payments_dataset.csv")
reviews = pd.read_csv("../data/olist_order_reviews_dataset.csv")
items = pd.read_csv("../data/olist_order_items_dataset.csv")
sellers = pd.read_csv("../data/olist_sellers_dataset.csv")
```

```python
customers_clean = customers.copy()
orders_clean = orders.copy()
products_clean = products.copy()
payments_clean = payments.copy()
reviews_clean = reviews.copy()
items_clean = items.copy()
sellers_clean = sellers.copy()
```

```python
# Check missing values again
products_clean.isnull().sum()
```

```python
# Replace missing categories with Unknown
products_clean["product_category_name"] = (
    products_clean["product_category_name"]
    .fillna("Unknown")
)
```

```python
# Verify category missing values
products_clean["product_category_name"].isnull().sum()
```

```python
# Count each order status
orders_clean["order_status"].value_counts()
```

```python
# Show cancelled orders
orders_clean[
    orders_clean["order_status"] == "canceled"
].head()
```

```python
# Convert order dates to datetime
orders_clean["order_purchase_timestamp"] = pd.to_datetime(
    orders_clean["order_purchase_timestamp"]
)

orders_clean["order_approved_at"] = pd.to_datetime(
    orders_clean["order_approved_at"]
)

orders_clean["order_delivered_customer_date"] = pd.to_datetime(
    orders_clean["order_delivered_customer_date"]
)

orders_clean["order_estimated_delivery_date"] = pd.to_datetime(
    orders_clean["order_estimated_delivery_date"]
)

```

```python
# Verify data types
orders_clean.info()
```

```python
# Extract purchase year
orders_clean["purchase_year"] = (
    orders_clean["order_purchase_timestamp"].dt.year
)
```

```python
# Extract purchase month
orders_clean["purchase_month"] = (
    orders_clean["order_purchase_timestamp"].dt.month
)
```

```python
# Extract month name
orders_clean["month_name"] = (
    orders_clean["order_purchase_timestamp"].dt.month_name()
)
```

```python
orders_clean[
    ["purchase_year", "purchase_month", "month_name"]
].head()
```

```python
# Calculate delivery days
orders_clean["delivery_days"] = (
    orders_clean["order_delivered_customer_date"]
    - orders_clean["order_purchase_timestamp"]
).dt.days
```

```python
orders_clean["delivery_days"].describe()
```

```python
# Save cleaned datasets

customers_clean.to_csv(
    "../data/cleaned/customers_clean.csv",
    index=False
)

orders_clean.to_csv(
    "../data/cleaned/orders_clean.csv",
    index=False
)

products_clean.to_csv(
    "../data/cleaned/products_clean.csv",
    index=False
)

print("Cleaned files saved successfully")
```
