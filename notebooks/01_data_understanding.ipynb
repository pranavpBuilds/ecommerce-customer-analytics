# Project Objective

Analyze e-commerce customer, order, payment, review, and product data to identify sales trends, customer behavior, payment preferences, seller performance, and factors affecting customer satisfaction.

# Dataset Overview

The Olist E-commerce dataset contains information about:

- Customers
- Orders
- Products
- Payments
- Reviews
- Sellers

The dataset represents a real Brazilian e-commerce marketplace and allows end-to-end business analysis.

# Initial Findings

- No duplicate rows were found.
- Some missing values exist in Orders and Products tables.
- Missing delivery dates are expected for cancelled or undelivered orders.
- Product metadata is incomplete for a small number of products.

```python
import pandas as pd
import numpy as np
```

```python
customers = pd.read_csv("../data/olist_customers_dataset.csv")
orders = pd.read_csv("../data/olist_orders_dataset.csv")
products = pd.read_csv("../data/olist_products_dataset.csv")
payments = pd.read_csv("../data/olist_order_payments_dataset.csv")
reviews = pd.read_csv("../data/olist_order_reviews_dataset.csv")
items = pd.read_csv("../data/olist_order_items_dataset.csv")
sellers = pd.read_csv("../data/olist_sellers_dataset.csv")

print("Datasets Loaded Successfully")
```

```python
datasets = {
    "Customers": customers,
    "Orders": orders,
    "Products": products,
    "Payments": payments,
    "Reviews": reviews,
    "Items": items,
    "Sellers": sellers
}

for name, df in datasets.items():
    print(f"{name}: {df.shape}")
```

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(df.columns.tolist())
```

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(df.info())
```

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(df.isnull().sum())
```

```python
for name, df in datasets.items():
    missing = df.isnull().sum()
    missing = missing[missing > 0]

    print(f"\n{name}")

    if len(missing) == 0:
        print("No missing values")
    else:
        print(missing)
```

```python
for name, df in datasets.items():
    print(f"{name}: {df.duplicated().sum()}")
```
